```python
def scrollDown(elemento, numberOfScrollDowns):
    while numberOfScrollDowns >=0:
        elemento.send_keys(Keys.PAGE_DOWN)
        numberOfScrollDowns -= 1
        time.sleep(3)
    #return browser
```
