```python
driver = webdriver.Chrome(executable_path='/home/florian/lab/project-mining-web-ml/chromedriver')
driver.get("https://www.google.com/maps/search/starbucks/@-12.054562,-77.0642815,14z/data=!3m1!4b1");
#driver.get("https://elcomercio.pe/deporte-total/futbol-mundial/barcelona-vs-getafe-vivo-espn-2-28-septiembre-2019-laliga-santander-horarios-canales-tv-movistar-laliga-movistar-mitele-plus-como-cuando-donde-ver-futbol-gratis-online-coliseum-alfonso-perez-lionel-messi-es-pe-nnda-nnrt-noticia/");
```

```python
def scrollDown(elemento, numberOfScrollDowns):
    while numberOfScrollDowns >=0:
        elemento.send_keys(Keys.PAGE_DOWN)
        numberOfScrollDowns -= 1
        time.sleep(3)
    #return browser
```

```python
page = 0
i = 0
#init counter
items = len(driver.find_elements_by_xpath("//div[@class='section-result']"))
#items = 2
while(True):
    try:    
        #init counter
        if(items == 0):
            i = 0
            items = driver.find_elements_by_xpath("//div[@class='section-result']")
            #items = 2
            
        while(items > 0):
            driver.find_elements_by_xpath("//div[@class='section-result']")[i].click()
            time.sleep(10)
            #time
            driver.get(driver.current_url)
            time.sleep(10)
            driver.find_element_by_class_name('allxGeDnJMl__button').click()

            #time
            time.sleep(10)
            driver.get(driver.current_url)
            time.sleep(10)
            #time
            elemento = driver.find_element_by_xpath("//div[contains(@class,'section-layout section-scrollbox scrollable-y scrollable-show')]")
            scrollDown(elemento, 3)

            soup = BeautifulSoup(driver.page_source)
            elem = [item.text for item in soup.find_all('span', {'class' : 'section-review-text'})]
            stars = [item.get("aria-label") for item in soup.find_all('span', {'class' : 'section-review-stars'})]
            frame = pd.DataFrame({'stars' : stars, 'comment' : elem})
            frame.to_csv("Deportes_tituares2.csv" + str(i) + "-" + str(page), index= False)
            driver.get("https://www.google.com/maps/search/starbucks/@-12.054562,-77.0642815,14z/data=!3m1!4b1")
            
            time.sleep(10)#time
            
            #click pagination button
            for k in range(0, page):
                driver.find_element_by_id('n7lv7yjyC35__section-pagination-button-next').click()
                time.sleep(10)
            
            items = items - 1
            i = i + 1
        
        #increment paganation
        page = page + 1
        
        #click pagination button
        for k in range(0, page):
            driver.find_element_by_id('n7lv7yjyC35__section-pagination-button-next').click()
            time.sleep(10)
        
    except OSError as err:
        print("error ->", err)
        break
    except NoSuchElementException as err:
        print("error ->", err)
        break
    except ElementNotVisibleException as err:
        ##exception for page empty
        print("error ->", err)
        break
    
```
